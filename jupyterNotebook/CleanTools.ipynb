```python
import pandas as pd
import numpy as np

# library for text preprocessing
import re
import nltk

#nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
#nltk.download('wordnet') 


from nltk.stem.wordnet import WordNetLemmatizer
```

```python
# selected column headers we want to use 
new_headers = ['Tags', 'Score']
df = pd.read_csv("ml2017.csv", header=0)
# remove unwanted columns keep selected headers
dataset = pd.DataFrame(df, columns=new_headers)

#dataset
```

```python
import re
# convert to lowercase
dataset['Tags'] = dataset['Tags'].str.lower()

# convert desc to string
dataset['Tags'] = dataset['Tags'].astype(str)

dataset['New_Tags'] = dataset['Tags'].apply(lambda x: ''.join([item for item in re.sub(r'.', '', x, count = 1)]))

dataset.head(10)
```

```python
product_headers = ['New_Tags', 'Score']
export_dataset = pd.DataFrame(dataset, columns=product_headers)
 
export_dataset
export_dataset.to_csv('Cml2017.csv', encoding='utf-8', index=False)
```

```python
# selected column headers we want to use 
new_headers = ['New_Tags', 'Score']
df = pd.read_csv("Cml2017.csv", header=0)
# remove unwanted columns keep selected headers
dataset = pd.DataFrame(df, columns=new_headers)

dataset
```

```python
# convert desc to string
dataset['New_Tags'] = dataset['New_Tags'].astype(str)

dataset['Tags'] = dataset['New_Tags'].apply(lambda x: ''.join([item for item in x.split('>')]))

dataset.head(10)
```

```python
product_headers = ['Tags', 'Score']
export_dataset = pd.DataFrame(dataset, columns=product_headers)
 
export_dataset
export_dataset.to_csv('Fml2017.csv', encoding='utf-8', index=False)
```

```python
import pandas as pd
import random
import os
import numpy as np
from tqdm import tqdm_notebook
from collections import Counter

from gensim.models import Word2Vec
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 10.0)
plt.rcParams.update({'font.size': 15})
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
```

```python
path = 'Fml2017.csv'
df = pd.read_csv(path)
df.columns
df.head()
```

```python
tags = list(df['Tags'].unique())
all_tags = []
for tag in tqdm_notebook(tags):
    all_tags += tag.split('<')
```

```python
count_tags = Counter(all_tags)
tags = pd.DataFrame([list(count_tags.keys()), list(count_tags.values())])
tags = tags.transpose()
tags.columns = ['tag', 'count']
tags = tags.sort_values(by='count', ascending=False)
```

```python
ml = ['dbn', 'convolutional-neural-network', 'conv-neural-network', 'genetic-programming', 'k-means', 'linear-regression', 'svm', 'logistic-regression', 'random-forest', 'recurrent-neural-network', 'rnn', 'decision-tree', 'genetic-algorithm', 'knn', 'naivebayes', 'bayesian-networks', 'hidden-markov-models', 'pattern-recognition', 'ensemble-learning']
x = []
y = []
for tag in ml:
    index = list(tags['tag']).index(tag)
    x.append(tag)
    y.append(list(tags['count'])[index])
    
plt.xticks(rotation='vertical')
plt.bar(x, y)
addlabels(x, y)
plt.ylabel('Tag')
plt.xlabel('Tag Raw-Count')
plt.title('Number of ML & DL Tags in Security 2017')
plt.show()
```

```python
m = ['genetic-programming', 'genetic-algorithm', 'k-means', 'svm', 'linear-regression', 'logistic-regression', 'random-forest', 'recurrent-neural-network', 'rnn', 'decision-tree', 'knn', 'naivebayes', 'hidden-markov-models', 'pattern-recognition', 'ensemble-learning']
x = []
y = []
for tag in m:
    index = list(tags['tag']).index(tag)
    x.append(tag)
    y.append(list(tags['count'])[index])
    
plt.xticks(rotation='vertical')
plt.bar(x, y)
addlabels(x, y)
plt.ylabel('Tag')
plt.xlabel('Tag Raw-Count')
plt.title('Number of ML Tags in Security 2017')
plt.show()
```

```python
import pandas as pd
import random
import os
import numpy as np
from tqdm import tqdm_notebook
from collections import Counter

from gensim.models import Word2Vec
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 10.0)
plt.rcParams.update({'font.size': 15})
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
```

```python
path = 'Ftools2017.csv'
df = pd.read_csv(path)
df.columns
df.head()
```

```python
tags = list(df['Tags'].unique())
all_tags = []
for tag in tqdm_notebook(tags):
    all_tags += tag.split('<')
```

```python
count_tags = Counter(all_tags)
tags = pd.DataFrame([list(count_tags.keys()), list(count_tags.values())])
tags = tags.transpose()
tags.columns = ['tag', 'count']
tags = tags.sort_values(by='count', ascending=False)
```

```python
tools = ['snort', 'splunk', 'suricata', 'aide', 'fail2ban', 'datadog', 'graylog', 'qradar', 'mcafee', 'tor', 'nmap', 'metasploit', 'kali-linux', 'wireshark', 'norton', 'tcpdump', 'aircrack-ng', 'openvas', 'nessus', 'sqlmap', 'zap', 'hydra']
x = []
y = []
for tag in tools:
    index = list(tags['tag']).index(tag)
    x.append(tag)
    y.append(list(tags['count'])[index])
    

plt.xticks(rotation='vertical')
plt.bar(x, y)
addlabels(x, y)
plt.ylabel('Tag')
plt.xlabel('Tag Raw-Count')
plt.title('Number of commonly used security tools in 2017')
plt.show()
```

```python
ids = ['suricata', 'fail2ban', 'snort', 'aide']
x = []
y = []
for tag in ids:
    index = list(tags['tag']).index(tag)
    x.append(tag)
    y.append(list(tags['count'])[index])
    
plt.bar(x, y)
addlabels(x, y)
plt.ylabel('Tag')
plt.xlabel('Tag Raw-Count')
plt.title('Security tools that fall under the IDS category in 2017')
plt.show()
```

```python
ips = ['datadog', 'fail2ban', 'splunk']
x = []
y = []
for tag in ips:
    index = list(tags['tag']).index(tag)
    x.append(tag)
    y.append(list(tags['count'])[index])
    
plt.bar(x, y)
addlabels(x, y)
plt.ylabel('Tag')
plt.xlabel('Tag Raw-Count')
plt.title('Security tools that fall under the IPS category in 2017')
plt.show()
```

```python
########################################################
# STEP 6 - Set up data for Queries and Graphs
########################################################

# Import data from job ad sites 
jobDF = pd.DataFrame(columns = ['Name', 'Link', 'Advertiser', 'Description'])
jobSites = ['Seek', 'Dice', 'Jora', 'SAJobSearch']
for site in jobSites:
    df = pd.read_csv('JobCrawl_' + site + '.csv', index_col=None, usecols=['Name', 'Link', 'Advertiser', 'Description'])
    df['Site'] = site
    jobDF = jobDF.append(df)
jobDF = jobDF.reset_index()

# Remove deliminated text from SAJobSearch data
for index, row in jobDF.iterrows():
    jobDF.at[index,'Description'] = jobDF.iloc[index].Description.replace('\n', ' ').replace('\r', ' ')

# Language names as keywords
keywords = [
    "C/C++",
    "C++",
    "C",
    "Java",
    "JavaScript",
    "Python",
    "R",
    "Ruby"
]
keywords = [x.lower() for x in keywords]
keywords_freq = dict.fromkeys(keywords, 0)


def count_keywords(total_words):
    return [w for w in total_words if w in keywords]


def count_keywords(total_words, keyword_list):
    return [w for w in total_words if w in keyword_list]


# The following two functions are missing from the code in the supplied report by Shubham Gupta, but appear to be sourced from : https://programminghistorian.org/en/lessons/counting-frequencies
def wordListToFreqDict(wordlist):
    wordfreq = [wordlist.count(p) for p in wordlist]
    return dict(list(zip(wordlist,wordfreq)))


# checks job description against a single dimensional list 
for all_info in jobDF.Description:
    total_words = all_info.lower().split()
    wordList = count_keywords(total_words, keywords)
    dictionary = wordListToFreqDict(wordList)
    for key in dictionary.keys():
        keywords_freq[key] += dictionary[key]
        
# Merge the "C/C++", "C++", and "C" results back into one 
keywords_freq['c/c++'] += keywords_freq['c++'] + keywords_freq['c']
keywords_freq.pop('c++', None)
keywords_freq.pop('c', None)

# Sort them 
keywords_freq = dict(sorted(keywords_freq.items(), key=lambda item: item[1], reverse=True))

# Get Stack Overflow data
dTypes = {'Language': object}
QuestionLibs = pd.read_csv('SO_BDA_Q_Libraries_in_Skill_Space.csv', dtype=dTypes)
dTypes = {'Language': object}
AnswerLibs = pd.read_csv('SO_BDA_A_Libraries_in_Skill_Space.csv', dtype=dTypes)
```

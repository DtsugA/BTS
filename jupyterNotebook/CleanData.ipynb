```python
# convert to lowercase
dataset['Description'] = dataset['Description'].str.lower()

# convert desc to string
dataset['Description'] = dataset['Description'].astype(str)

#dataset.dtypes

# create a list of stop words and adding custom stopwords
stop_words = set(stopwords.words("english"))

# create a list of custom stopwords
new_words =  ['job', 'title', 'click','apply', 'years', 'status', 'team','matter', 'have', 'you', 'must', '•', 'on', 'work','other', 'we', 'or', 'that', '', 'our', 'will', 'are', 'an', 'by', 'all', '-', 'for', 'this', 'can', 'it', 'with', 'good', 'over', '100%', 'so', 'is', 'be', 'the', 'and', 'his','to', 'in', 'a', 'he', 'at', 'of', 'from', 'was', 'as']
stop_words = stop_words.union(new_words)

#stop_words

dataset['New_Description'] = dataset['Description'].apply(lambda x: ' '.join([item for item in x.split() if item not in stop_words]))

dataset.head(10)
```

```python
# Fetch wordcount for each description
dataset['word_count'] = dataset['New_Description'].apply(lambda x: len(str(x).split(" ")))
dataset[['New_Description','word_count']].head()

dataset
```
